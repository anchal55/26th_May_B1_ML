```python
x = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms','Area Population']]
y = df['Price']
```

# 1. splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size=0.25, random_state=10)
```

```python
xtrain.info()
```

```python
xtest.info()
```

```python
ytrain.shape
```

```python
ytest.shape
```

```python
xtrain.head()
```

# 2. creating the machine learning modl

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

# 3. training the model with training data

```python
model.fit(xtrain,ytrain)
```

```python
model.coef_
```

```python
pd.DataFrame(model.coef_, columns=['Coef.'],index=x.columns)    #relation between price and othr columns means if 14 income increase so price would be increase 21.591947
```

# 4. evaluation 

```python
print('%.1f%%'%(model.score(xtest,ytest)*100))
```

```python
#this accuracy
```

```python
ypred= model.predict(xtest)
```

```python
ypred
```

```python
ytest
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(ytest,ypred))
```
